```python
import torch

# La fonction magique !
cuda_available = torch.cuda.is_available()

print(f"PyTorch : CUDA disponible ? {cuda_available}")

if cuda_available:
    print(f"Nom du GPU : {torch.cuda.get_device_name(0)}")
    print(f"Version CUDA utilisée par PyTorch : {torch.version.cuda}")
    print(f"Nombre de GPU disponibles : {torch.cuda.device_count()}")
else:
    print("PyTorch ne détecte pas CUDA. Vérifiez votre installation.")
```
